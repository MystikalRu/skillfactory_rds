# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
#data.groupby('production_companies')['title_length'].mean().sort_values(ascending=False).head(2)
#pd.DataFrame(data.production_companies.str.split('|').tolist()).stack()
def counter(movie_bd, x):
    data_plot=movie_bd[x].str.cat(sep='|')
    dat=pd.Series(data_plot.split('|'))
    info=dat.value_counts(ascending=False)
    return info
sum_gen = counter(data,'production_companies')
for gen in sum_gen.index:
    sum_gen[gen] = data['title_length'][data['production_companies'].map(lambda x: True if gen in x else False)].mean()
sum_gen = pd.DataFrame(sum_gen).sort_values(0, ascending=False)
print(sum_gen)
answer_ls.append('5.Four By Two Productions')
```

# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
c = Counter() 
c1 = Counter() 
c3 = Counter() 
for index, row in data.iterrows(): 
    names = names = row['director'].split('|') 
    for name in names: 
        c[name] += 1 
        c1[name] += 0 
        if(row['cost']>0): 
            c1[name] += 1 
        c3[name] = c1[name]/c[name] 
c3.most_common()

answer_ls.append(4.Christopher Nolan)
```
